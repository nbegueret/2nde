## <center>**Vocabulaire et définitions :**

**Dioptre** : surface de séparation exemple dioptre air-eau = surface de séparation entre l’air et l’eau

**Point d’incidence I** : point de la surface qui est atteint par le rayon lumineux.

**Normale à la surface**: droite perpendiculaire à la surface au niveau du point d’incidence.

**Rayon incident** : rayon qui arrive sur le dioptre

**Rayon réfracté** : rayon qui a traversé le dioptre 

**Rayon réfléchi**:  rayon qui part du dioptre sans l’avoir traversé

**Angle d’incidence i1** : angle que forment le rayon incident et la normale à la surface.

**Angle de réfraction i2** : angle que forment le rayon réfracté et la normale à la surface.

**Document 1 : Conditions d’invisibilité d’un tube à essais**

Un tube à essais rempli d’eau plongé dans un bécher, contenant lui aussi de l’eau, est visible.
Dans cette situation, comme le matériau constituant le tube n’a pas le même indice de réfraction que l’eau,
la lumière est  déviée par les parois du tube. 
Cette lumière déviée à la surface de séparation entre les deux milieux pénètre alors dans nos yeux : le tube est visible.
En revanche, si les rayons lumineux ne sont pas réfractés lorsqu’ils changent de milieu, alors le tube est invisible.


**Document 2 : Détermination expérimentale de l’indice de réfraction**

Pour déterminer l’indice de réfraction d’un milieu matériel transparent, on peut utiliser le dispositif ci-dessous.
On fait varier l’angle d’incidence i1 et on mesure à chaque fois l’angle de réfraction i2. 
On rassemble les résultats dans un tableau de mesures et on utilise la loi de Snell-Descartes pour la réfraction afin d’en déduire l’indice de réfraction n2 du milieu transparent.

3eme Loi de Snell- Descartes :  
$$ n_1×sin⁡(i_1)  =n_2  ×sin(⁡i_2)  $$

Comme les angles ont en degré, il faut vérifier le paramétrage de la calculatrice. 
Les angles sont en degré ; on accède à la fonction sinus via la calculatrice. 


Un indice de réfraction n’a pas d’unité. 

Exemples

indice de réfraction de l’air  :	$n(air) = 1,0 $

indice de réfraction du cyclohexane : 	$n(cyclohexane) = 1,4 $

Indice de réfraction de l’éthanol : 	$n(éthanol) = 1,35$

Indice de réfraction du glycérol : 	$n(glycérol) = 1,5$

Indice de réfraction de diiodométhane : 	$n(diiodométhane) = 1,9$

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
#on rentre les valeurs expériementales trouvées au cours de l'expérience. 
i1=[0,10]
i2=[0,7]

#on cherche le modèle linéaire correspondant
a,_,_,_,_=linregress(i1,i2)
x_modele=np.linspace(0,max(i1),101) #permet de créer une liste de 101 valeurs d'abscisses
y_modele=[a*i for i in x_modele] #calcul l'ordonnée en appliquant le modèle

# on trace la i2 en fonction de i1
plt.plot(i1,i2,'or' )
plt.plot(x_modele, y_modele) 

plt.axis([0,90,0,90])
plt.grid()
plt.show()
```
